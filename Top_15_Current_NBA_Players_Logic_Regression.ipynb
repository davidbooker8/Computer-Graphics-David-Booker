```python
import pandas as pd
# Data Set Top 15 Current NBA Players
file_path = '/Top 15 Current Basketball Players.csv'
nba_data = pd.read_csv(file_path)

#Display first couple of rows of the dataset to confirm it has loaded correctly
print(nba_data.head())
```

```python
from sklearn.preprocessing import LabelEncoder

#Encode the 'Position' column
le = LabelEncoder()
nba_data['Position_encoded'] = le.fit_transform(nba_data['Position'])

#Need to Define features and targeted variables for this script
X = nba_data[['Age', 'Year Started', 'Points Per Game', 'Championships', 'All-NBA', 'Position_encoded']]
y = nba_data['MVPS']. apply(lambda x: 1 if x > 0 else 0) #Binary target : 1 if MVP > 0, else 0
```

```python
from sklearn.model_selection import train_test_split

#Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

#Initialize and train the logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

#Make predicitions on the test set
y_pred = model.predict(X_test)

#Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100: .2f}%")

print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd

#Get the coefficients of the logistics regression model

coefficients = pd.DataFrame({'Feature': X.columns, 'Coefficients': model.coef_[0]})
print(coefficients)
```
