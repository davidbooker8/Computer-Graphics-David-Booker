```python
import PIL.Image
import PIL.ImageDraw
import face_recognition
```

```python
#load the jpg file into a numpy array
image = face_recognition.load_image_file("IMG_20231001_104327.jpg")
```

```python
# find all the face in the image
face_locations = face_recognition.face_locations(image)
```

```python
number_of_faces = len(face_locations)
print("I found {} face(s) in this photograph.".format(number_of_faces))
```

```python
#Load image into python image library obj for the ability to draw on top and display
pil_image = PIL.Image.fromarray(image)
```

```python
#Print the location of each image
for face_location in face_locations:

    top, right, bottom, left = face_location
    print("A face is located at pixel location Top: {}, Left: {}, Bottom: {}, Right: {}".format(top, left, bottom, right))
draw = PIL.ImageDraw.Draw(pil_image)
draw.rectangle([left, top, right, bottom], outline="green")
```

```python
import matplotlib.image as mpimg
from matplotlib.pyplot import imshow

%matplotlib inline
from IPython.display import Image

imshow(pil_image)
```

```python
#Find facial features
face_landmarks_list = face_recognition.face_landmarks(image)
```

```python
# Loop over each face
for face_landmarks in face_landmarks_list:

  #Loop over each facial feature (eye, nose, mouth, lips, etc)
  for name, list_of_points in face_landmarks.items():

    #Print the location of each facial feature
    print("The {} in this face has the following points: {}".format(name, list_of_points))

    #Trace out each facial feature with a line
    draw.line(list_of_points, fill="yellow", width=2)
```

```python
# Generate the face encoding
face_encodings = face_recognition.face_encodings(image)
```

```python
#display face encoding
if len(face_encodings) == 0:
  #No faces found in the image
  print("No faces were found")
else:
  # Grab the first face encoding
  first_face_encoding = face_encodings[0]

  #Print the result
  print(first_face_encoding)
```
