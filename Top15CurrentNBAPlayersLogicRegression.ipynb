```python
# Loading and checking to see the data structure of the dataset
import pandas as pd

# Load the CSV file
file_path = '/content/Top 15 Current Basketball Players.csv'
nba_data = pd.read_csv(file_path)

# Display some of the rows of the data to verify that it is working
nba_data.head()

```

```python
# Preprocessing the data to perform a logic regression
from sklearn.preprocessing import LabelEncoder

#Encode 'Position' column
le = LabelEncoder()
nba_data['Position_encoded'] = le.fit_transform(nba_data['Position'])

#Defining Feature and target variable
X = nba_data[['Age', 'Year Started', 'Points Per Game', 'Championships', 'All-NBA', 'Position_encoded']]
y = nba_data['MVPS'].apply(lambda x: 1 if x > 0 else 0) # Binary target: 1 if MVP > 0 else 0
```

```python
# Split dataset into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

#Create and train the logic regression model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

#Make predictions on the test set
y_pred = model.predict(X_test)

#Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

#Print Classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd

coefficients = pd.DataFrame({'Feature': X.columns, 'Coefficient': model.coef_[0]})
print(coefficients)
```
