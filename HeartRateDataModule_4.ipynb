```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
from sklearn.linear_model import LinearRegression
#Read the Data
data = pd.read_csv('heartratedata - Sheet1.csv')

data['Heart Attack'] = data['Heart Attack'].apply(lambda x: 1 if x == 'Y' else 0)

#Split data into X and Y
X = data[['Heart Rate']]
y = data['Heart Attack']

#Create a LR model
linear_model = LinearRegression()
linear_model.fit(X, y)

#Predict the model
y_pred = linear_model.predict(X)

#Plot scatter plot of and the regression line
plt.figure(figsize=(20.0, 10.0))
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred, color='red', label='Regression line')


plt.xlabel('Heart Rate')
plt.ylabel('Heart Attack')
plt.title('Heart Rate vs Heart Attack')
plt.legend()

#Display
plt.show()

#Output weights
print('Intercept (b0):', linear_model.intercept_)
print('Coefficient (b1):', linear_model.coef_[0])
```

```python
from sklearn.linear_model import LogisticRegression
#Create LR Model
logistic_model = LogisticRegression()
logistic_model.fit(X, y)

#Probability of a heart attack for a heart rate =60
new_input = np.array([[60]]) #New heart rate input
probability = logistic_model.predict_proba(new_input)

#Output for 60 heart rate
print(f"Probability of heart attack for heart rate of 60: {probability[0][1]:.2f}")

#Output LR weights
print("Intercept (b0):", logistic_model.intercept_)
print("coefficients (b1)", logistic_model.coef_[0])
```
