```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
#Iris dataset Load
data = pd.read_csv('iris.csv')
iris_df = pd.DataFrame(data)
iris_df.head()
iris_df.info()

```

```python
#Summarize information and place in scatterplot
iris_df.describe()
sns.pairplot(iris_df, hue='class', markers=["o", "s", "D"])
plt.show()

#Show each distribution of the features
iris_df.hist(edgecolor='black', figsize=(10,8))
plt.tight_layout()
plt.show()

#Compare feaure distribution across the classes
plt.figure(figsize=(12,6))
sns.boxplot(data=iris_df, x='class', y='sepal_length')
plt.show()
```

```python
#Define matrix and target variable
x= iris_df.drop(columns='class')
y= iris_df['class']

#Split dataset into training and test set
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=42, stratify=y)
```

```python
#Standard Data
scaler=StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)

#KNN Model
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)
y_pred = knn.predict(X_test_scaled)

#View the results
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=iris_df['class'].unique(), yticklabels=iris_df['class'].unique())
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```
